```python
import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# 1. Criando o DataFrame
# ===============================

data = {
    'Nome': ['Alice', 'Joao', 'Charlie', 'David', 'Eva', 'Diego', 'Denize', 'Claudio'],
    'Idade': [25, 30, 35, 40, 45, 60, 22, 24],
    'Profissão': ['Engenheiro', 'Médico', 'Professor', 'Advogado', 'Médico','Engenheiro', 'Estudante','Estudante'],
    'Salário': ['4500', '8000', '5000', '10000', '12000','15000', '1200','1500'],
    'Limite_Credito': ['2500', '4000', '4000', '1000', '10000','2000', '500','250'],
    'Historico_Inadimplencia': ['0', '0', '0', '1', '0','1', '0','1'],
    'Estado_Civil': ['Casamento', 'Casamento', 'Solteiro', 'Solteiro', 'Casamento','Solteiro', 'Solteiro','Solteiro'],
    'Imovel_Proprio': ['0', '0', '0', '1', '1','1', '0','0']
}

df = pd.DataFrame(data)

# Convertendo colunas numéricas
df["Salário"] = df["Salário"].astype(float)
df["Limite_Credito"] = df["Limite_Credito"].astype(float)
df["Historico_Inadimplencia"] = df["Historico_Inadimplencia"].astype(int)
df["Imovel_Proprio"] = df["Imovel_Proprio"].astype(int)

df.head()
```

```python
plt.figure(figsize=(10,6))
plt.scatter(df["Salário"], df["Limite_Credito"], s=100)
plt.xlabel("Salário", fontsize=12)
plt.ylabel("Limite de Crédito", fontsize=12)
plt.title("Relação entre Salário e Limite de Crédito", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(10,6))
df.groupby("Profissão")["Limite_Credito"].mean().plot(kind="bar")
plt.xlabel("Profissão", fontsize=12)
plt.ylabel("Limite de Crédito Médio", fontsize=12)
plt.title("Limite de Crédito Médio por Profissão", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

```python
plt.figure(figsize=(8,5))
df.groupby("Imovel_Proprio")["Limite_Credito"].mean().plot(kind="bar")
plt.xlabel("Imóvel Próprio (0 = Não, 1 = Sim)", fontsize=12)
plt.ylabel("Limite de Crédito Médio", fontsize=12)
plt.title("Impacto do Imóvel Próprio no Limite de Crédito", fontsize=14)
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```
